```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_loc = "../data/assigData2.tsv"
```

```python
bins_weight = 30 
bins_diameter = 20

alpha = 0.5

red = (1.0,0.0,0.0,alpha)
orange = (0.9,0.4,0.0,alpha)
purple = (0.7,0.0,1.0,alpha)
```

```python
dataset = pd.read_csv(data_loc, sep="	")
```

```python
ax_weight = dataset["W_apl"].hist(
    bins=bins_weight,
    color="red",
    histtype=u'step',
    figsize=(12,5)
    )
dataset["W_apl"].hist(
    bins=bins_weight,
    legend=True,
    color=red,
    ax=ax_weight
    )
dataset["W_orng"].hist(
    bins=bins_weight,
    legend=True,
    color=orange,
    ax=ax_weight)
dataset["W_orng"].hist(
    bins=bins_weight,
    color="orange",
    histtype=u'step',
    ax=ax_weight)
dataset["W_grp"].hist(
    bins=bins_weight,
    legend=True,
    color=purple,
    ax=ax_weight)
dataset["W_grp"].hist(
    bins=bins_weight,
    color="purple",
    histtype=u'step',
    ax=ax_weight)

ax_weight.set_title("Weight Histogram")
ax_weight.set_ylabel("Count")
ax_weight.set_xlabel("Weight")
```

```python
ax_diameter = dataset["D_apl"].hist(
    bins=bins_diameter,
    color="red",
    histtype=u'step',
    figsize=(12,5)
    )
dataset["D_apl"].hist(
    bins=bins_diameter,
    legend=True,
    color=red,
    )
dataset["D_orng"].hist(
    bins=bins_diameter,
    legend=True,
    color=orange,
    ax=ax_diameter)
dataset["D_orng"].hist(
    bins=bins_diameter,
    color="orange",
    histtype=u'step',
    ax=ax_diameter)
dataset["D_grp"].hist(
    bins=bins_diameter,
    legend=True,
    color=purple,
    ax=ax_diameter)
dataset["D_grp"].hist(
    bins=bins_diameter,
    color="purple",
    histtype=u'step',
    ax=ax_diameter)

ax_diameter.set_title("Diameter Histogram")
ax_diameter.set_ylabel("Count")
ax_diameter.set_xlabel("Diameter")
```

There is an overlap between the weights of apple and oranges. A classifier will not be able to distingush by weight for apple and oranges.

There is lower ovelap when considering diameter. 

Hence, select diameter.
